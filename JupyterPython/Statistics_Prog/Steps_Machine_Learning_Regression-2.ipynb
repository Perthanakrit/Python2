# K-fold approach to evaluate the performance

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits=5,shuffle=False)
```

```python
y=y.to_numpy()
```

```python
y
```

```python
X=X.to_numpy()
```

```python
rmse_train = 0
rmse_test =0
for train,test in kf.split(X):
    lr=linear_model.LinearRegression()
    reg=lr.fit(X[train],y[train])
    y_pred_train =reg.predict(X[train])
    y_pred_test =reg.predict(X[test])
    e_train=y[train]-y_pred_train
    e_test = y[test]-y_pred_test
    #rmse_train += np.sqrt(np.mean(e_train*e_train))     # compute rmse for the estimation rmse train data
    #rmse_test += np.sqrt(np.mean(e_test*e_test))     # compute rmse for the estimation rmse train data
    rmse_train  += np.sqrt( metrics.mean_squared_error(y[train],y_pred_train))
    rmse_test  += np.sqrt( metrics.mean_squared_error(y[test],y_pred_test))
    
rmse_train_5cv = rmse_train/5            #average the rmse
rmse_test_5cv = rmse_test/5            #average the rmse



print('RMSE from the train data on 5-fold CV: {}'.format(rmse_train_5cv))


print('RMSE from the test data on 5-fold CV: {}'.format(rmse_test_5cv))

```

```python
rmse_train = 0
rmse_test =0
for train,test in kf.split(X):
    lr=linear_model.Lasso()
    reg=lr.fit(X[train],y[train])
    y_pred_train =reg.predict(X[train])
    y_pred_test =reg.predict(X[test])
    e_train=y[train]-y_pred_train
    e_test = y[test]-y_pred_test
    #rmse_train += np.sqrt(np.mean(e_train*e_train))     # compute rmse for the estimation rmse train data
    #rmse_test += np.sqrt(np.mean(e_test*e_test))     # compute rmse for the estimation rmse train data
    rmse_train  += np.sqrt( metrics.mean_squared_error(y[train],y_pred_train))
    rmse_test  += np.sqrt( metrics.mean_squared_error(y[test],y_pred_test))
    
rmse_train_5cv = rmse_train/5            #average the rmse
rmse_test_5cv = rmse_test/5            #average the rmse



print('RMSE from the train data on 5-fold CV: {}'.format(rmse_train_5cv))


print('RMSE from the test data on 5-fold CV: {}'.format(rmse_test_5cv))

```

```python
rmse_train = 0
rmse_test =0
for train,test in kf.split(X):
    lr=linear_model.Ridge()
    reg=lr.fit(X[train],y[train])
    y_pred_train =reg.predict(X[train])
    y_pred_test =reg.predict(X[test])
    e_train=y[train]-y_pred_train
    e_test = y[test]-y_pred_test
    #rmse_train += np.sqrt(np.mean(e_train*e_train))     # compute rmse for the estimation rmse train data
    #rmse_test += np.sqrt(np.mean(e_test*e_test))     # compute rmse for the estimation rmse train data
    rmse_train  += np.sqrt( metrics.mean_squared_error(y[train],y_pred_train))
    rmse_test  += np.sqrt( metrics.mean_squared_error(y[test],y_pred_test))
    
rmse_train_5cv = rmse_train/5            #average the rmse
rmse_test_5cv = rmse_test/5            #average the rmse



print('RMSE from the train data on 5-fold CV: {}'.format(rmse_train_5cv))


print('RMSE from the test data on 5-fold CV: {}'.format(rmse_test_5cv))

```
