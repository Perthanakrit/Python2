# Classification
- use DecisionTreeClassifier (sklearn.tree)

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np

x = [[27],[35], [39], [21], [22], [45]]

y_name = ["lemon", "orange", "orange", "lemon", "lemon"  , "orange"]
y = [0, 1, 1, 0, 0, 1]
```

```python
from IPython.display import Image
```

Data Visualization

```python
plt.xlabel("fruit")
plt.ylabel("size")
plt.scatter(y_name, x)
plt.show()
```

Modal

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```

```python
model.fit(x,y)
```

Predict

```python
x_input = [[25.3]]
y_predict = model.predict(x_input)
y_predict
```

```python
fruit_classes = ["Lemon", "Orange"]

prd = fruit_classes[y_predict[0]]
print(prd)
```

```python
x_input = [[31.0],
           [25.0],
           [21.2],
           [38.6]]

y_predict = model.predict(x_input)
y_predict
```

```python
print("- Result -")
print("{:>6} {:>8}".format("size", "fruit"))

for i, val in enumerate(y_predict):
    print(f"size: {x_input[i]} = {val}  fruit = {fruit_classes[val]}")
```

Visualize Model

```python
from sklearn.tree import plot_tree

data_feature_names = ["size"]

plt.subplots(nrows=1, ncols=1, figsize=(4, 4), dpi=96)

plot_tree(model,
          filled=True, rounded=True,
          feature_names=data_feature_names, class_names=fruit_classes)

plt.show()
```

 ขอบเขตการตัดสินใจ (Descisioi regions)

```python
from mlxtend.plotting import plot_decision_regions
```

```python
x = np.array(x)
y = np.array(y)
print(x)
print(y)
```

Classification 2: (Fruit 3 classes)

```python
#Dataset
size = [[27],
        [35], [39], [21], [22], [45], [102],[110],[120]]

class_name = ['lemon', 'orange', 'orange', 'lemon',
               'lemon', 'orange', 'pamelo', 'pamelo', 'pamelo']

y_class = [0, 1, 1, 0, 0, 1, 2, 2, 2]
```

```python
plt.xlabel("fruit")
plt.ylabel("size")

plt.scatter(class_name, size)
plt.show()
```

Model

```python
model = DecisionTreeClassifier()
model.fit(size, y_class)
```

```python
fruit_classes = ["Lemon", "Orange", "Pamelo"] # 0, 1, 2
```

Predict

```python
size_input = [[102]]
y_predict = model.predict(size_input)

prd = fruit_classes[y_predict[0]]

print(f"Result: {prd}")
```

```python
size_input = [[28.0], [32], [39], [181]]

y_predict = model.predict(size_input)

for i, k in enumerate(y_predict):
    print(f"size: {size_input[i]} = {fruit_classes[k]}")
```

```python
from sklearn.tree import plot_tree

data_feature_names = ["size"]

plt.subplots(nrows=1, ncols=1, figsize=(5, 4), dpi=96)

plot_tree(model,
          filled=True, rounded=True,
          feature_names=data_feature_names, class_names=fruit_classes)

plt.show()
```
