-- 2 Feature 2 Classes --
> Ex. ผลไม้ใช้ size mass ช่วยclassify 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns # scatter plot
```

# Dataset

```python
df = pd.read_csv('../data/fruit2class.csv')
df
```

# Data Visualization

```python
plt.scatter(df.m, df.d) # X = mass, Y = diameter(d)

plt.show()
```

```python
sns.scatterplot(data=df, x="m", y="d", hue="target", style="target", s=100)

plt.show()
```

# Model

```python
x = df.drop(['target'], axis=1) # นำ target(class) ออกจาก X 
#X.head()
x
```

```python
y = df.target # นำ target มาเก็บไว้ใน y
#y.head()
y
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier() # สร้างโมเดล

model.fit(x, y) # train โมเดล
```

# Predict

```python
x_input = [[5.6, 3.5]] # diameter = 5.6, mass = 3.5 / 2 Features (m, d)

y_predict = model.predict(x_input) # ทำนายผล
y_predict
```

```python
fruit_class = ['Grape', 'Lemon']
```

```python
x_input = [[5.6, 3.5],
           [7.8, 4.4],
           [5.9, 3.0]]

y_predict = model.predict(x_input)

for i, val in enumerate(y_predict):
    print(f"{x_input[i]}: {fruit_class[val]}")
```

# Model Visualization
- Tree
- ขอบเขต

```python
from sklearn.tree import plot_tree

data_frame_name = ["m", "d"]

plt.subplots(nrows=1, ncols=1, figsize=(8,7), dpi=96)

plot_tree(model, feature_names=data_frame_name, class_names=fruit_class, filled=True, rounded=True)

plt.show()
```

# Decision regions

```python
from mlxtend.plotting import plot_decision_regions
```

```python
x = np.array(x)
y = np.array(y)
```

```python
plot_decision_regions(x, y, clf=model)

plt.xlabel('mass')
plt.ylabel('diameter')

plt.show()
```

```python
plot_decision_regions(x, y, clf=model)

x_input = [[5.6, 4.2],
           [7.5, 4.5],
           [5.9, 3.0]]

plt.scatter(x_input[0][0], x_input[0][1], s=200, c='r')
plt.scatter(x_input[1][0], x_input[1][1], s=200, c='b')
plt.scatter(x_input[2][0], x_input[2][1], s=200, c='g')

plt.xlabel('mass')
plt.ylabel('diameter')

plt.show()
```

## Classification 2: (Fruit 3 classes)

- 2 feature 3 classes 

ex. grape, lemon, orange

# Dataset

```python
df_3_class = pd.read_csv('../data/fruit3class.csv')
df_3_class
```

# Data Visualization

```python
sns.scatterplot(data=df_3_class, x="m", y="d", hue="target", style="target", s=100, palette="Set1")

plt.show()
```

# Model

```python
feature = df_3_class.drop(['target'], axis=1) #axis1 => column
#feature.head()
feature
```

```python
y = df_3_class.target
y
```

```python
model = DecisionTreeClassifier(max_depth=5)
model.fit(feature, y)
```

```python
fruit_3_classes = ['Grape', 'Lemon', 'Orange']
```

# Predict

```python
feature_input = [[5.6, 4.2],
                 [7.6, 4.5],
                 [7.8, 5.4]]

y_predict = model.predict(feature_input)
y_predict
```

```python
for i, val in enumerate(y_predict):
    print(f"{feature_input[i]}: {fruit_3_classes[val]}")
```

# Model Visualize

```python
from sklearn.tree import plot_tree

data_frame_name = ["m", "d"]

plt.subplots(nrows=1, ncols=1, figsize=(8, 7), dpi=96)

plot_tree(model, feature_names=data_frame_name,
          class_names=fruit_3_classes, filled=True, rounded=True)

plt.show()
```

# Decision regions

```python
from mlxtend.plotting import plot_decision_regions
```

```python
ax = plot_decision_regions(feature, y, clf=model)

plt.xlabel('mass')
plt.ylabel('diameter')
plt.title('Decision Tree')

handles, labels = ax.get_legend_handles_labels()

ax.legend(handles, fruit_3_classes, framealpha=0.3, scatterpoints=1) # framealpha คือ ความโปร่งใสของกรอบ

plt.show()
```
